## Exercice 1 (Lasso seuillé)

### Chargement des données:

```python
from sklearn.datasets.mldata import fetch_mldata
dataset_name = 'leukemia'
data = fetch_mldata(dataset_name)
X = data.data
y = data.target
X = X.astype(float)
y = y.astype(float)
```

### Chargement des librairies:

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import (manifold, datasets, decomposition, ensemble,
                     discriminant_analysis, random_projection)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LassoCV
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
```

### 1) Nombre d'observations et de features:

```python
nb_obs, nb_features = X.shape[0], X.shape[1]
print('Le nombre d\'observations est de %d et le nombre de features est de %d' % (nb_obs, nb_features))
```

p >> n, une selection et/ou retraitement des variables est nécessaire.

Normalisons les variables explicatives:

```python
scaler = StandardScaler(with_mean=False).fit(X)
X_stand = scaler.transform(X)
```

### 2) ACP et TSNE:

```python
# Initialisation de PCA avec 1 et 2 composantes principales
pca_1 = PCA(n_components=1)
pca_2 = PCA(n_components=2)

# Fit de la PCA
X_pca_1 = pca_1.fit_transform(X_stand)
X_pca_2 = pca_2.fit_transform(X_stand)


# Initialisation de la TSNE avec 1 et 2 composantes principales
tsne_1 = manifold.TSNE(n_components=1, init='pca', random_state=0)
tsne_2 = manifold.TSNE(n_components=2, init='pca', random_state=0)

# Fit tsne
X_tsne_1 = tsne_1.fit_transform(X)
X_tsne_2 = tsne_2.fit_transform(X)
```

Visualisons les points obetenus en fonction du label y

```python
fig = plt.figure(1)

# Plot de PCA avec 1 axe
plt.subplot(221)
for k in (-1, 1):
    Xk_pca_1 = X_pca_1[y == k]
    plt.plot(Xk_pca_1[:, 0], np.zeros_like(Xk_pca_1), '.', markersize=8, label=k, alpha=0.8)
    plt.legend(numpoints=1,loc=1)
plt.tight_layout()
plt.title("PCA en une composante")

# Plot de PCA avec 2 axes
plt.subplot(223)
for k in (-1, 1):
    Xk_pca_2 = X_pca_2[y == k]
    plt.plot(Xk_pca_2[:, 0], Xk_pca_2[:, 1], '.', markersize=8, label=k, alpha=0.8)
    plt.legend(numpoints=1,loc=1)
plt.tight_layout()
plt.title("PCA en deux composantes")

# Plot de TSNE avec 1 axe
plt.subplot(222)
for k in (-1, 1):
    Xk_tsne_1 = X_tsne_1[y == k]
    plt.plot(Xk_tsne_1[:, 0], np.zeros_like(Xk_tsne_1), '.', markersize=8, label=k, alpha=0.8)
    plt.legend(numpoints=1,loc=1)
plt.tight_layout()
plt.title("TSNE en une composante")

# Plot de TSNE avec 2 axes
plt.subplot(224)
for k in (-1, 1):
    Xk_tsne_2 = X_tsne_2[y == k]
    plt.plot(Xk_tsne_2[:, 0], Xk_tsne_2[:, 1], '.', markersize=8, label=k, alpha=0.8)
    plt.legend(numpoints=1,loc=1)
plt.tight_layout()
plt.title("TSNE en deux composantes")

plt.show()
```

Le découpage en 1 ou 2 composantes principales ne semble pas suffisant pour récupérer les labels y. Toutefois, la TSNE en deux composantes semble approcher une meilleure classification avec les 1 et les -1 séparés grossièrement par l'axe des abscisses (remarquons aussi que les echelles de la tsne sont beaucoup plus grandes et donc distingue mieux les différents points).

### 3) Constituer un train set et un test set:

On utilise train_test_split de sklearn pour constituer les test et train set:

```python
X_train, X_test, y_train, y_test = train_test_split(X_stand, y, test_size=0.2, random_state=42)
```

(X_test constitué à partir de X_stand, mais peut-être aurait-il dû être constitué du fit sur X_train une fois ce dernier standardisé séparemment)

### 4) Trouver mathématiquement une relation entre lambda' et lambda:

D'après la CNO, on a:
 1/n Xt(XÔ - y) + lambda'Ô = 0, soit Ô = inv((XtX + nlambda'))Xty d'où lambda' = 1/n*lambda

### 5) Utiliser LassoCV:

```python
# initialiser les paramètres
cv = 4
n_alphas = 17
delta = 2
alpha_0 = np.max(np.dot(np.transpose(X_train), y_train)) / X_train.shape[0]
alphas = [alpha_0 * 10**(-(delta * k / (n_alphas - 1))) for k in range(n_alphas)]

# initialiser le model
lassoCV = LassoCV(n_alphas=n_alphas, alphas=alphas, cv=cv, fit_intercept=False)

# fit du model sur les données
lassoCV_train = lassoCV.fit(X_train, y_train)

# evaluer le modèle sur le test set
y_test_pred = lassoCV_train.predict(X_test)
lassoCV_mse = mean_squared_error(y_test, y_test_pred, multioutput="uniform_average")
print("L'erreur quadratique moyenne est: ", lassoCV_mse)
```

```python
# Calculer la mse moyenne pour toutes les cv sur chaque valeur de lambda'
alphas_mseMean = np.mean(lassoCV_train.mse_path_, axis=1)

# plotter les mse en fonction des lambdas'
fig = plt.figure()
plt.semilogx(alphas, alphas_mseMean, label='MSE')
plt.legend()
plt.tight_layout()
plt.title("MSE en fonction des lambdas'")
plt.show()
```

### 6) Proposer un estimateur sigma du bruit du modèle:

L'erreur quadratique normalisée est un estimateur biaisé du bruit que nous pouvons choisir, soit le mse.

```python
def sigma(X, y):
    y_pred = lassoCV_train.predict(X)
    return mean_squared_error(y, y_pred, multioutput="uniform_average")
```

### 7) Coder le lasso seuillé:

```python
class lasso_seuil(LinearRegression):
    
    def __init__(self, alpha=1, tau=0.1):
        self.alpha = alpha
        self.tau = tau
        self.intercept_ = 0.
    
    def fit(self, X, y):
        # initialiser le model
        # Lasso simple pour éviter la cross-val à cette étape
        lasso = Lasso(alpha=self.alpha, fit_intercept=False)

        # fit du model sur les données
        lasso_train = lasso.fit(X, y)

        # récupérer les coefficients du lasso
        thetas = lasso_train.coef_

        # initialiser la liste d'indice
        S = []

        # test sur les coefficients
        for j, theta in enumerate(thetas):
            if abs(theta) > self.tau:
                S += [j]

        # initialiser l'ols
        ols = LinearRegression(fit_intercept=False, normalize=False)

        # fitter ols
        #self = ols.fit(X[:, S], y)

        # selectionner sur le support       
        X = X[:, S]
        
        # récupérer l'ouput
        self.coef_, self._residues, self.rank_, self.singular_ = \
                ols.fit(X, y).coef_, ols.fit(X, y)._residues, ols.fit(X, y).rank_, ols.fit(X, y).singular_
        self.coef_ = self.coef_.T
        
        # récupérer le support
        self.support = S

        return self
    
    def predict(self, X):
        
        X = X[:, self.support]
        
        return self._decision_function(X)
```

### 8) Coder une procédure de cross-validation sur lambda et tau:

```python
# Initialiser les paramètre à tester
params = [{'alpha': alphas, 'tau': [0.001, 0.005, 0.01, 0.02, 0.05]}]
score = "neg_mean_squared_error"

# lancer la procédure de recherche du meilleur couple de paramètres
clf = GridSearchCV(lasso_seuil(), params, cv=4, scoring=score)
```

```python
clf.fit(X_train, y_train)
```

```python
# calculer le meilleur score (score etait neg_mean_square_error)
lasso_seuil_score = - clf.best_score_
y_lasseuil_pred = clf.predict(X_test)
lasseuil_mse = mean_squared_error(y_test, y_lasseuil_pred, multioutput="uniform_average")
print('Le mse de la lasso seuillé est:', lasseuil_mse)
```

### 9) Comparer les résultats:

Pour comparer les modèles, il nous ne manque plus le modèle des estimateurs des moindres carrées:

```python
# initialiser le model
ols = LinearRegression(fit_intercept=False, normalize=False)

# fit du model sur les données
ols_train = ols.fit(X_train, y_train)

# evaluer le modèle sur le test set
y_test_pred = ols_train.predict(X_test)
ols_mse = mean_squared_error(y_test, y_test_pred, multioutput="uniform_average")
print ('le mse pour OLS sans intercept est:', ols_mse)
```

De ce calcul et des question 8 et 5, il apparaît que le lassoCV est la meilleure méthode au regard de la mse.
